```python
import sklearn
import pickle
import pandas as pd

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import LinearSVC

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
sentiment_data = pd.read_csv('datasets/sentimental_analysis_data.csv', 
                             header=None,
                             names=['Label', 'Text'],
                             sep='\t')

sentiment_data.sample(10)
```

```python
sentiment_data.shape
```

```python
X = sentiment_data['Text']

Y = sentiment_data['Label']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
tfidf_vect = TfidfVectorizer(max_features=15)
```

```python
logistic_clf = LogisticRegression(solver='liblinear')
```

```python
clf_pipeline = Pipeline(steps=[('tfidf_vect', tfidf_vect), ('classifier', logistic_clf)])

pipeline_model = clf_pipeline.fit(x_train, y_train)
```

```python
y_pred = pipeline_model.predict(x_test)
```

```python
acc_score =  accuracy_score(y_test, y_pred)

acc_score
```

```python
pickle.dump(pipeline_model, open('models/logistic_regression_clf/model.pkl', 'wb'))
```

```python
decistion_tree_clf = DecisionTreeClassifier(max_depth=10)
```

```python
clf_pipelines = Pipeline(steps=[('tfidf_vect', tfidf_vect), ('classifier', decistion_tree_clf)])

pipeline_model = clf_pipelines.fit(x_train, y_train)
```

```python
y_pred = pipeline_model.predict(x_test)
```

```python
acc_score =  accuracy_score(y_test, y_pred)

acc_score
```

```python
pickle.dump(pipeline_model, open('models/decision_tree_clf/model.pkl', 'wb'))
```

```python
linear_svc_clf = LinearSVC(C=1.0, max_iter=100)
```

```python
clf_pipelines = Pipeline(steps=[('tfidf_vect', tfidf_vect), ('classifier', linear_svc_clf)])

pipeline_model = clf_pipelines.fit(x_train, y_train)
```

```python
y_pred = pipeline_model.predict(x_test)
```

```python
acc_score = accuracy_score(y_test, y_pred)

acc_score
```

```python
pickle.dump(pipeline_model, open('models/linear_svc_clf/model.pkl', 'wb'))
```
